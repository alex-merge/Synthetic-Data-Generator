# Using SDG to generate organoid like datasets

## Importing SDG

```python
from SDG import SDG

dataset = SDG()
```

## Looking at the documentation

```python
help(SDG)
```

## Creating a very basic dataset rotating on the X-axis only

```python
import random

origin = [random.randrange(-100, 101) for k in range(3)]
radius = 40
sample = 100 # Sample per time points
tx, ty, tz = 15, 10, -20

dataset.addRotatingSphere(origin = origin, radius = radius, sample = sample, variation = False, theta_x = tx, theta_y = ty,
                         theta_z = tz)
dataset.data.head()
```

## Looking the data
Showing the points for the third and fourth time point.

```python
dataset.showData(TP = 3); dataset.showData(TP = 4)
```

Showing the path of the "track" 20.

```python
dataset.showPath(track = 20)
```

```python
savepath = r"D:\Wrk\Presentations\September\datasets\S1"

dataset.exportCSV(savepath, OAT = True)
```
